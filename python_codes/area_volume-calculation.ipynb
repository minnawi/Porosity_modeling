# <span style="color:green"> Script for caculating the surface area of the pore-bone interface and the volumen of the pore
   

# Load libraries

```python
import os
import pandas as pd
import trimesh
```

### Can be changed

```python
# Directory containing STL files
stl_dir = "calculations"                   # replace with your folder path
excel_file = "calculations/area-volume.xlsx"  # results file
```

### Calculation on all STL files in that directory

```python
# Check if Excel exists, otherwise create it
if os.path.exists(excel_file):
    df = pd.read_excel(excel_file)
else:
    df = pd.DataFrame(columns=["sample", "A", "Vpore"])

# Loop through STL files in the directory
for filename in os.listdir(stl_dir):
    if filename.lower().endswith(".stl"):
        stl_path = os.path.join(stl_dir, filename)
        mesh = trimesh.load_mesh(stl_path)

        # Calculate surface area and volume
        A = mesh.area
        Vpore = mesh.volume

        # Extract sample name
        sample_name = os.path.splitext(filename)[0]

        # Append new row
        new_row = {"sample": sample_name, "A": A, "Vpore": Vpore}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

        print(f"Processed {sample_name}: A={A:.3f}, Vpore={Vpore:.3f}")

# Save results back to Excel
df.to_excel(excel_file, index=False)
print(f"\nAll results saved to {excel_file}")
```
