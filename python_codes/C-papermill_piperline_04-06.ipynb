```python
 def run_processing_workflow(sample_dir):

    # A- Run `05_create_high-quality_model.ipynb`
    paper_dict = dict(nrrd_file = sample_dir,
                      thresholding_method = thresholding_method,
                      method_1 = method_1,
                      method_2 = method_2,
                      void_max = void_max,
                      hexa_min = hexa_min,
                      cylinder_height = cylinder_height,
                      cylinder_radius = cylinder_radius,
                      markup = markup,
                     )
    
    input_nb = '05_create_high-quality_model.ipynb'
    output_nb = str(parentpath) + '/' + input_nb
    
    print(input_nb)
    pm.execute_notebook(input_nb, output_nb, parameters=dict(paper_dict));

    

     # B- Run `06-Cleaning&repairing.ipynb`
    paper_dict = dict(filepath = parentpath
                     )
    
    input_nb = '06-Cleaning&repairing.ipynb'
    output_nb = str(parentpath) + '/' + input_nb

    print(output_nb)
    pm.execute_notebook(input_nb, output_nb, parameters=dict(paper_dict));
```

## <span style="color:RED"> Samples

```python
# Specify sample directory and side name
sample_dir = 'tail5/CEP_4T/segmented_volumes/full/microCT_scan.nrrd'

# PARAMETERS
# if thresholding_method = manual → assign threshold values for void_max and hexa_min ( at 1.2- Build cylinder model)
# if thresholding_method = automatic → choose a method [HUANG/INTERMODES/ ISO_DATA/KITTLER_ILLINGWORTH/LI/MAXIMUM_ENTROPY/MOMENTS/OTSU/RENYI_ENTROPY/SHANBHAG/TRIANGLE/YEN]
thresholding_method = "manual" # "automatic" or "manual"
method_1 = 'MOMENTS' #for automatic thresholding
method_2 = 'SHANBHAG' #for automatic thresholdingCEP_1B
void_max = 9729 #for manual thresholding
hexa_min = 11010 #for manual thresholding

# Dimensions of the cylindrical ROI
cylinder_height = 1 #mmCEP_1B
cylinder_radius = 0.5 #mm

# Define the different markup values
markup_list = ['left_L', 'mid_L', 'right_L']

# Loop through each markup value
path = Path(sample_dir)
for markup in markup_list:
    # Define the parent path
    parentpath = os.path.join(path.parent.absolute(), markup)
    # Check if the directory exists, and create it if it doesn't
    if not os.path.exists(parentpath):
        os.makedirs(parentpath)
    run_processing_workflow(sample_dir)

```
